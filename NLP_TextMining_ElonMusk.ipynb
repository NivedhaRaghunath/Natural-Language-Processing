```python
stemmer = PorterStemmer()

def preprocessing(text):
    
    text = text.lower()
    text = re.sub("[^a-z:) ]" ," ",text)
    text = text.split()
    text = [stemmer.stem(i) for i in text if i not in stop_words ]
    text = " ".join(text)
    return text
```

```python
y = []

for i in data["Text"]:
    i_lower = i.lower().strip()
    j_lower = j.lower().strip() # Convert to lowercase and remove leading/trailing whitespace
    if any(j_lower in i_lower for j in positive_words):
        y.append(1)
    elif any(j_lower in i_lower for j in negative_words):
        y.append(0)
    else:
        y.append("neutral")  # Default to 0 if no words are found





```
